```python
# Data Cleaning (removing missing values)
```

```python
#turning M for Malignant and B for Benign Tumor to binary
df = pd.get_dummies(df)
df = df.drop('diagnosis_B',axis=1)
```

```python

df_for_svm = df[['symmetry_worst','perimeter_worst']]
```

```python
df_for_svm
```

```python
#Applying Machine Learning Model: Support Vector Machine
```

```python
y = df['diagnosis_M']
X = df_for_svm
y
X
```

```python
y
```

```python
#Train | Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVR, LinearSVR
```

```python
#potential kernel options for Support Vector Machines are
#  -> linear
#  -> rbf : radial basis function, consider C and Gamma
#  -> sigmoid
#  -> poly : considder C and degree
base_model = SVR()
base_model.fit(scaled_X_train, y_train)
```

```python
base_preds = base_model.predict(scaled_X_test)
```

```python
#Evaluating (MAE, MSE)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(y_test,base_preds)
```

```python
np.sqrt(mean_squared_error(y_test, base_preds))
```

```python
y_test.mean()
```

```python
# Grid Search for Better Model
```

```python
param_grid = {'C':[0.001,0.01, 0.1, 0.5, 1, 10],
                 'kernel' :['linear','rbf','poly'],
                 'gamma': ['scale','auto'],
                 'degree' : [2,3,4],
                 'epsilon': [0,0.01,0.1,0.5,1,2,3]}
    
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svr = SVR()
grid = GridSearchCV(svr,param_grid=param_grid)
```

```python
grid.fit(scaled_X_train, y_train)
```

```python
grid.best_params_
```

```python
grid_preds = grid.predict(scaled_X_test)
```

```python
mean_absolute_error(y_test, grid_preds)
```

```python
np.sqrt(mean_squared_error(y_test,grid_preds))
```
